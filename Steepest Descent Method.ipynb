```
import sympy as sym
import numpy as np
from sympy.utilities.lambdify import lambdify, implemented_function
```

```
x,y,d = sym.symbols('x,y,d')
z = x-y+(2*(x**2))+(2*x*y)+(y**2)
iterations = 2
s,r = 0,0
dx = sym.diff(z,x)
dy = sym.diff(z,y)
print('dx: ' + str(dx))
print('dy: ' + str(dy))
for i in range(iterations):
  print()
  print('k: ' + str(i+1) + '    x: (' + str(s) + ',' + str(r) + ')')
  lam_f = lambdify((x,y), [dx,dy])
  e1,e2=lam_f(s,r)
  s1 = -1*e1;  s2 = -1*e2
  print('s: (' + str(s1) + ',' + str(s2) + ')')
  u1 = s+d*s1;  u2 = r+d*s2
  print('univariate points: (' + str(u1) + ',' + str(u2) + ')')
  lam_z = lambdify((x,y),z)
  fun = lam_z(u1,u2)
  print('univariate function: ' + str(fun))
  dd = sym.diff(fun,d)
  print('dd: ' + str(dd))
  a = sym.solve(sym.Eq(dd,0),d)
  print('lambda: ' + str(a))
  s = s + a[0]*s1; r = r + a[0]*s2
  print('new points: (' + str(s) + ',' + str(r) + ')')
lam_q = lambdify((x,y),z)
value = lam_z(s,r)
print()
print('Function value at (' + str(s) + ',' + str(r) + ') is ' + str(value))
```
