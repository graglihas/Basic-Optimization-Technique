```
def f(x):
  return (100-x)**2
```

## Exhaustive Search Method

```
#Interval
a=0; b=5

#number of function evaluation = n+1
e=11
n=e-1

#lgorithm
x1=a
dx=(b-a)/n
x2=x1+dx; x3=x2+dx
flag=0
while x3<=b:
  y1=f(x1); y2=f(x2); y3=f(x3)
  if y1>=y2 and y2<=y3:
    print('\nThe minimum point lies between (' + str(x1) + ',' + str(x3) +')')
    flag=1
    break
  else:
    x1=x2; x2=x3; x3=x2+dx
if flag==0:
  print('No minimum exists between (' + str(a) + ',' + str(b) +')')
```

## Bounding Phase Method

```
#Initial Guess
x0 = 30

#Increment
dx = 5
k=0

if f(x0-dx)>=f(x0) and f(x0)>=f(x0+dx):
  dx=dx
if f(x0-dx)<=f(x0) and f(x0)<=f(x0+dx):
  dx=-1*dx
x1=x0
x2=x1+(pow(2,k)*dx)
print(f(x1))
print(f(x2))
while f(x2)<f(x1):
  print(k)
  k=k+1
  x0=x1
  x1=x2
  x2=x1+(pow(2,k)*dx)
else:
  print('\nThe minimum point lies between (' + str(x2) + ',' + str(x0) +')')
```

## Interval Halving method

```
#lower and upper bound
a=60;b=150
#accuracy
acc=0.25
xm=(a+b)/2
L=b-a
for i in range(3):
  x1=a+(L/4); x2=b-(L/4)
  if f(x1)<f(xm):
    b=xm
    xm=x1
  if f(x2)<f(xm):
    a=xm
    xm=x2
  if f(xm)<f(x1) and f(xm)<f(x2):
    a=x1
    b=x2
  L=b-a
print('\nThe minimum point lies between (' + str(a) + ',' + str(b) +')')
```

## Fibonacci Search

```
def fib(x):
  if x==0:
    return 1
  if x==1:
    return 1
  return fib(x-1)+fib(x-2)
```

```
#lower and upper bound
a=0;b=5
n=3
k=1
while k<n:
  k=k+1
  L=b-a
  L_=(fib(n-k+1)/fib(n+1))*L
  x1=a+L_
  x2=b-L_
  if f(x2)>f(x1):
    b=x2
  if f(x1)>f(x2):
    a=x1
print('\nThe minimum point lies between (' + str(a) + ',' + str(b) +')')
```
