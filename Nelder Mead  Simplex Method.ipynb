```
import math
import numpy as np
```

```
def fun(x):
  return x[0]-y+(2*(x**2))+(2*x*y)+(y**2)
  #return abs(math.sin(x[0])*(x[1]**3)+1)+(x[0]**2)+(0.1*(x[1]**4))
```

```
x=np.array([[0.,0.],[1.2,0.],[0.,0.8]])
y=np.array([0.,0.,0.])
d = 2
b = 0.5
iterations =3
print('delta: ' + str(d) + '    beta: ' + str(b) + '    iterations: ' + str(iterations))
for i in range(iterations):
  print()
  for j in range(3):
    y[j]=fun(x[j])
  print('x1: ' + str(x[0]) + '    x2: ' + str(x[1]) + '   x3: ' + str(x[2]))
  print('y1: ' + str(y[0]) + '    y2: ' + str(y[1]) + '   y3: ' + str(y[2]))
  h = int(np.where(y == np.max(y))[0])
  l = int(np.where(y == np.min(y))[0])
  g = 3-h-l
  print('h: ' + str(x[h]) + '    g:' + str(x[g]) + '   l: ' + str(x[l]))
  x_c = (x[g]+x[l])/2
  print('x_c: ' + str(x_c))
  x_r = 2*x_c-x[h]
  y_r = fun(x_r)
  print('x_r: ' + str(x_r) + '    y_r: ' + str(y_r))
  if y_r < y[l]:
    x_new = ((1+d)*x_c)-(d*x[h])
  if y_r >= y[h]:
    x_new = ((1-b)*x_c)+(b*x[h])
  if y_r > y[l] and y_r < y[h]:
    x_new = ((1+b)*x_c)-(b*x[h])
  print('x_new: ' + str(x_new))
  x[h] = x_new
```
