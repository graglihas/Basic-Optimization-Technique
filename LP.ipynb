```
from scipy.optimize import linprog as lp
```

```
co = [3,8]
Ai = [[-1,-1],[1,0],[0,-1]]
Bi = [-200,80,-60]
Ae = [[0,1,-1],[1,1,1]]
Be = [0,1]
method= 'simplex'
#method = 'revised simplex'
```

```
lp(c=co,A_ub=Ai,b_ub=Bi,method=method)
```

```
lp(c=co,A_ub=Ai,b_ub=Bi,A_eq=Ae,b_eq=Be,method=method)
```

```
lp(c=co,A_eq=Ae,b_eq=Be,method='interior-point',options={'maxiter':1})
```
