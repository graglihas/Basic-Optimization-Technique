```
import sympy as sym
import numpy as np
from sympy.utilities.lambdify import lambdify, implemented_function
```

```
x,y = sym.symbols('x,y')
#z = (2*(x**3)) + (3*(y**2)) + (3*y*(x**2)) - (24*y)
z = (((x**2)+(y**2)-1)**2)+(((y**2)-1)**2)
dx = sym.diff(z,x)
dy=sym.diff(z,y)
print('dx: ' + str(dx))
print('dy: ' + str(dy))
f = sym.Eq(dx,0)
g = sym.Eq(dy,0)
a = np.asarray(sym.solve([f,g],[x,y]))
print('extreme points: ' + str(a.tolist()))
dxdx = sym.diff(dx,x)
dxdy = sym.diff(dy,x)
dydx = sym.diff(dx,y)
dydy = sym.diff(dy,y)
print('dxdx: ' + str(dxdx))
print('dxdy: ' + str(dxdy))
print('dydx: ' + str(dydx))
print('dydy: ' + str(dydy))
print('EP      | H1 | H2')
for i in range(a.shape[0]):
  s,r=a[i].tolist()
  lam_f = lambdify((x,y), [dxdx,dxdy,dydx,dydy])
  e1,e2,e3,e4=lam_f(s,r)
  H1 = e1
  H2 = e1*e4 - e2*e3
  print(str(a[i].tolist()) + '| ' + str(H1) + ' | ' + str(H2))
```
